```python
from math import *
class Point:
    #constructor
    def __init__(self, coordx=0,coordy=0):
        self.x=coordx
        self.y=coordy
    def move(self,dx,dy):
        self.x=self.x+dx
        self.y=self.y+dy
    def distPoint(self,pt):
        dist=sqrt((self.x-pt.x)**2+(self.y-pt.y)**2)
        return dist
    def affichePoint(self):
        print(f"({self.x}, {self.y})")
p1 = Point(1, 1)
p2 = Point(2, 2)
p1.affichePoint()
p2.affichePoint()
print(p1.distPoint(p2))
type(p1)
```
