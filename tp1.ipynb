```python
#Exercice 1
from math import *
class CompteBancaire:
    def __init__(self, nomBanque, nom, solde):
        #V2rification du nom de la banque
        if not nomBanque.strip():
            raise ValueError("le nom de la banque ne peut pas être vide")
        #Vérification du nom du titulaire
        if not nom.strip():
            raise ValueError("le nom du titulaire ne peut pas être vide")
        #Vérification du solde
        if solde < 0:
            raise ValueError("le solde ne peut pas être négatif")
        #Initialistion des attribus
        self.nomBanque = nomBanque
        self.nom = nom
        self.solde = solde

    def deposer(self,montant):
        if montant is None:
            raise ValueError("le Montant ne peut pas être null")
        if montant < 0:
            raise ValueError("le Montant ne peut pas être négatif")
            
        self.solde += montant
    def retirer(self,montant):
        if montant < 0:
            raise ValueError("Le Montant ne peut pas être négatif")
        if montant > self.solde:
            raise ValueError("le Solde est insufisant")
        self.solde -= montant
    def afficher(self):
        print(f"Banque : {self.nomBanque}")
        print(f"Titulaire : {self.nom}")
        print(f"Solde : {self.solde}")
        # Création des comptes
c1 = CompteBancaire("Banque A", "Yassine", 1000)
c2 = CompteBancaire("Banque B", "Ahmed", 2000)
c3 = CompteBancaire("Banque C", "Sara", 1500)

listComptes = [c1, c2, c3]

# Tri par ordre alphabétique du nom du titulaire
listComptes.sort(key=lambda compte: compte.nom)
for compte in listComptes:
    compte.afficher()
```

```python
#Exercice 2
from math import *
class Date:
    @classmethod
    def estBissextile(cls, y):
        return (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0)
    def __init__(self,y,m,d):
        if m < 1 or m > 12:
            raise ValueError ("Le mois est compris dans l’intervalle [1, 12]")
        if y < 0:
            raise ValueError ("L’année doit être strictement positive")
        # Nombre de jours de chaque mois
        nb_jours = [31, 28, 31, 30, 31, 30,
                    31, 31, 30, 31, 30, 31]
        
        if m == 2 and Date.estBissextile(y):
            nb_jours[1] = 29
            
        if d < 1 or d > nb_jours[m - 1]:
            raise ValueError("Jour invalide.")
        #initiation des attributs
        self.y = y
        self.m = m
        self.d = d
dt1 = Date(2024,1,1)
dt1.estBissextile(2024)
```
